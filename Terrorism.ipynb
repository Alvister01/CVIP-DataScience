# EXPLORATORY DATA ANALYSIS

ATTACKS PER YEAR

```python
attack_year = df['Year'].unique()
y_count_years = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (16,10))
sns.barplot(x = attack_year,
           y = y_count_years,
           palette = 'mako')
plt.xticks(rotation = 40)
plt.xlabel('Year of Attack')
plt.ylabel('Number of Attacks ')
plt.title('Number of Attacks Per Year')
plt.show()
```

```python
attacks_year = df['Year'].value_counts(dropna = False).sort_index().to_frame().reset_index().rename(columns = {'index' : 'Year','Year':'Attacks'})
attacks_year.head(5)
```

```python
# We can conlude that the least number of attacks were made in 1971 and the most number of attacks in 2014
```

REGION WISE ATTACKS

```python
region_attack = df['Region'].value_counts().head(10)
region_attack
```

```python
region_attack.plot(kind = 'bar', color = 'mediumpurple',figsize = (16,9))
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.title('Region Wise Attack')
plt.show()
```

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',stacked = False,figsize=(18,6))
plt.title('Region wise attacks in each Year')
plt.ylabel('Number of Attacks')
plt.xlabel('Year of Attack')
plt.show()
```

```python
# Most number of attacks happened after 2010
# Most number of attacks took placed in Middle East & North Africa
```

COUNTRY WISE ATTACKS

PEOPLE KILLED (REGION WISE)

```python
killed_region = df[['Region','kill']].groupby('Region').sum().sort_values(by = 'kill', ascending = False)
killed_region
```

```python
killed_region.plot(kind = 'bar', color = 'red',figsize = (16,9))
plt.xlabel('Regions')
plt.ylabel('Number of People Killed')
plt.title('People Killed Region Wise')
plt.show()
```

```python
Total_Killed = df.loc[:,'kill'].sum()
print(Total_Killed)
```

```python
# Most number of people were killed in Middle East & North Africa
# Least number of people were killed in Australasia & Oceanic 
```

```python
# Total Number of people killed was 411868
```
